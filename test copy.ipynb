```python
from diffusers import DiffusionPipeline
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'

pipeline = DiffusionPipeline.from_pretrained(
    'runwayml/stable-diffusion-v1-5', safety_checker=None)

scheduler = pipeline.scheduler
tokenizer = pipeline.tokenizer

del pipeline

device, scheduler, tokenizer
```

```python
%run encoder.ipynb
%run vae.ipynb
%run unet.ipynb


encoder.to(device)
vae.to(device)
unet.to(device)

encoder.eval()
vae.eval()
unet.eval()
```

```python
@torch.no_grad()
def generate(text,neg_text=''):
    #text to tensor
    #[1, 77]
    pos = tokenizer(text,
                    padding='max_length',
                    max_length=77,
                    truncation=True,
                    return_tensors='pt').input_ids.to(device)
    neg = tokenizer(neg_text,
                    padding='max_length',
                    max_length=77,
                    truncation=True,
                    return_tensors='pt').input_ids.to(device)

    #[1, 77, 768]
    pos = encoder(pos)
    neg = encoder(neg)

    #[1+1, 77, 768] -> [2, 77, 768]
    out_encoder = torch.cat((neg, pos), dim=0)

    #vae noise image
    out_vae = torch.randn(1, 4, 64, 64, device=device)

    #50 time steps
    scheduler.set_timesteps(50, device=device)
    for time in scheduler.timesteps:

        #add noise
        #[1+1, 4, 64, 64] -> [2, 4, 64, 64]
        noise = torch.cat((out_vae, out_vae), dim=0)
        noise = scheduler.scale_model_input(noise, time)

        #calc noise
        #[2, 4, 64, 64],[2, 77, 768],scala -> [2, 4, 64, 64]
        pred_noise = unet(out_vae=noise, out_encoder=out_encoder, time=time)

        #pos-neg
        #[2, 4, 64, 64] -> [1, 4, 64, 64]
        pred_noise = pred_noise[0] + 7.5 * (pred_noise[1] - pred_noise[0])

        #add more noise
        #[1, 4, 64, 64]
        out_vae = scheduler.step(pred_noise, time, out_vae).prev_sample

    #output to decoder
    out_vae = 1 / 0.18215 * out_vae
    #[1, 4, 64, 64] -> [1, 3, 512, 512]
    image = vae.decoder(out_vae)

    #into image
    image = image.cpu()
    image = (image + 1) / 2
    image = image.clamp(0, 1)
    image = image.permute(0, 2, 3, 1)
    return image.numpy()[0]


```

```python
import matplotlib.pyplot as plt
pic = generate('Bear laying down lifting weights')
plt.imshow(pic)
plt.show()
```

```python

def read_file_lines(filename):
    # Create an empty list to store the lines
    lines = []
    
    # Open the file in read mode
    with open(filename, 'r') as file:
        # Read each line in the file
        for line in file:
            # Strip newline characters and add to the list
            lines.append(line.strip())
    
    return lines
def gen(dir):
    line = read_file_lines(dir)
    for i in range(len(line)):
        if i != 1 :
            continue
        print(line[i])
        pic = generate(line[i])
        fig, ax = plt.subplots()
        ax.imshow(pic)
        ax.axis('off')

        ax.set_xticks([])  # No x ticks
        ax.set_yticks([])  # No y ticks

    # Remove padding and margins from the figure and save
        plt.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace=0)
        plt.savefig(f'./Snorlax_original_promt_1/snorlax_{i}.png', bbox_inches='tight', pad_inches=0, dpi=300)
        plt.close(fig)
        print(f'./Snorlax_original_promt_1/snorlax_{i}.png')
gen('./prompt/snorlax.txt')
```
